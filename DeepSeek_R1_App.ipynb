## Front End Application

```python
def generate_response(prompt_style, question, model_choice):
    """
    生成医学 AI 回答
    :param prompt_style: 选择的 prompt 模板
    :param question: 用户输入的医学问题
    :param model_choice: "Professional" 或 "Patient"
    :return: 结构化的医学回答
    """
    input_text = prompt_style.format(question, "")
    inputs = tokenizer([input_text], return_tensors="pt").to(model.device)

    if model_choice == "Patient":
       output = model.generate(**inputs,
                            max_new_tokens=1500,
                            do_sample=True,
                            top_k=20,
                            top_p=0.7,
                            temperature = 0.4)
    else:
        output = model.generate(**inputs,
                            max_new_tokens=1500,
                            do_sample=True,
                            top_k=5,
                            top_p=0.4,
                            temperature = 0.2)
    response = tokenizer.batch_decode(output, skip_special_tokens=False)
    response_text = response[0].split('### Response:')[1]

    # 去掉 <｜end▁of▁sentence｜> 特殊 token
    response_text = response_text.replace("<｜end▁of▁sentence｜>", "").strip()

    # 提取思维链 (CoT)
    cot_start = response_text.find("<think>") + len("<think>")
    cot_end = response_text.find("</think>")

    cot_reasoning = response_text[cot_start:cot_end].strip() if cot_end != -1 else "No CoT found"
    final_answer = response_text[cot_end + len("</think>"):].strip() if cot_end != -1 else response_text.strip()

    if model_choice == "Patient":
        return final_answer
    else:
        return f"""
    <div class='cot-text'>{cot_reasoning}</div>
    <div class='answer-text'>{final_answer}</div>
    """
```

```python
def generate_response_interface(question, model_choice):
    """
    用户接口：允许选择 Professional 或 Patient 风格
    :param question: 用户输入的问题
    :param model_choice: "Professional" 或 "Patient"
    :return: 结构化 AI 响应
    """
    professional_prompt_style = """Below is a clinical inquiry requiring a precise, evidence-based medical response.
Your task is to provide a structured, well-reasoned clinical analysis, integrating current guidelines, peer-reviewed literature, and established best practices.

Before formulating a response, critically appraise the clinical scenario, identifying key pathophysiological mechanisms, differential diagnoses, and appropriate investigative and therapeutic strategies.
Ensure that all recommendations adhere to contemporary clinical protocols and are substantiated by high-quality medical evidence.

### Instruction:
You are an advanced AI medical consultant specializing in differential diagnosis, clinical decision-making, and treatment optimization.
Your objective is to deliver a succinct yet comprehensive response tailored to healthcare professionals, prioritizing accuracy, clarity, and clinical applicability.

### Question:
{}

### Response:
<think>{}"""

    patient_prompt_style = """Below is an instruction that describes a task, paired with relevant background information.
Your task is to generate a well-reasoned and structured response.

Before providing an answer, carefully analyze the question and construct a logical, step-by-step reasoning process.
Make sure to justify your conclusions based on medical knowledge and clinical evidence.

### Instruction:
You are an advanced medical AI assistant with expertise in clinical reasoning, diagnostics, and treatment planning.
Your goal is to analyze the given medical question and provide a well-structured, evidence-based response.
You can assume you are helping a doctor.

### Question:
{}

### Response:
<think>{}"""

    if model_choice == "Professional":
        return generate_response(professional_prompt_style, question, model_choice)
    else:
        return generate_response(patient_prompt_style, question, model_choice)
```

```python
import time

def slowly_reverse(word, progress=gr.Progress()):
    progress(0, desc="Starting")
    time.sleep(1)
    progress(0.05)
    new_string = ""
    for letter in progress.tqdm(word, desc="Reversing"):
        time.sleep(0.25)
        new_string = letter + new_string
    return new_string
```

```python
def generate_response(prompt_style, question, model_choice):
    """
    生成医学 AI 回答
    :param prompt_style: 选择的 prompt 模板
    :param question: 用户输入的医学问题
    :param model_choice: "Professional" 或 "Patient"
    :return: 结构化的医学回答
    """
    input_text = prompt_style.format(question, "")
    inputs = tokenizer([input_text], return_tensors="pt").to(model.device)

    if model_choice == "Patient":
       output = model.generate(**inputs,
                            max_new_tokens=1500,
                            do_sample=True,
                            top_k=20,
                            top_p=0.7,
                            temperature = 0.4)
    else:
        output = model.generate(**inputs,
                            max_new_tokens=1500,
                            do_sample=True,
                            top_k=5,
                            top_p=0.4,
                            temperature = 0.2)
    response = tokenizer.batch_decode(output, skip_special_tokens=False)
    response_text = response[0].split('### Response:')[1]

    # 去掉 <｜end▁of▁sentence｜> 特殊 token
    response_text = response_text.replace("<｜end▁of▁sentence｜>", "").strip()

    # 提取思维链 (CoT)
    cot_start = response_text.find("<think>") + len("<think>")
    cot_end = response_text.find("</think>")

    cot_reasoning = response_text[cot_start:cot_end].strip() if cot_end != -1 else "No CoT found"
    final_answer = response_text[cot_end + len("</think>"):].strip() if cot_end != -1 else response_text.strip()
    final_answer = "Response: " + final_answer
    cot_reasoning = "Chain of Thought: " + cot_reasoning

    if model_choice == "Patient":
        return f"""
    <div class='answer-style'>
        {final_answer}
    </div>
    """
    else:
        return f"""
    <div class='cot-style'>
        {cot_reasoning}
    </div>
    <div class='answer-style'>
        {final_answer}
    </div>
    """
```
