```python
from huggingface_hub import HfApi

api = HfApi()

model_path = "/content/drive/MyDrive/285_project/DeepSeek-R1-Medical-CoT"  # 你的模型文件夹
repo_name = "zijiechen156/DeepSeek-R1-Medical-CoT"  # 你的 Hugging Face 仓库名，格式 "用户名/模型名"

# 创建新 repo（如果你还没创建）
api.create_repo(repo_id=repo_name, private=False)  # private=True 表示私有仓库

# 上传整个模型文件夹
api.upload_folder(
    folder_path=model_path,  # 本地模型路径
    repo_id=repo_name,       # Hugging Face 仓库
    repo_type="model"
)
```
